### yelp_academic_dataset_Business & Tip

```python
import pandas as pd

tips = pd.read_json('yelp_academic_dataset_tip.json', lines=True)
business = pd.read_json('yelp_academic_dataset_business.json', lines=True)
```

```python
business_tip = pd.merge(tips, business, on='business_id', how='inner')
```

```python
business_tip.info()
```

```python
# Filter rows that contain the keyword 'Restaurants'
restaurantsdf = business_tip[business_tip['categories'].str.contains("Restaurants", case=False, na=False)]
restaurantsdf = restaurantsdf.reset_index(drop=True)
```

```python
restaurantsdf.info()
```

```python
unique_businesses = restaurantsdf['business_id'].nunique()
unique_users = restaurantsdf['user_id'].nunique()
print(f"Unique Businesses: {unique_businesses}")
print(f"Unique Users: {unique_users}")
```

```python
# Filter the restaurants_data dataframe to include only rows where 'city' is 'Santa Barbara'
sbrestaurants = restaurantsdf[restaurantsdf['city'] == 'Santa Barbara']
```

```python
sbrestaurants.info()
```

```python
sorted_sbrestaurants = sbrestaurants.sort_values(by='name', ascending=False)
```

```python
sorted_sbrestaurants
```

```python
print(sbrestaurants)
```

```python
sb_unique_businesses = sbrestaurants['business_id'].nunique()
sb_unique_users = sbrestaurants['user_id'].nunique()
print(f"Unique Businesses: {sb_unique_businesses}")
print(f"Unique Users: {sb_unique_users}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Extract top 20 most common words
vec = CountVectorizer(stop_words='english', max_features=20)
word_counts = vec.fit_transform(sbrestaurants['text'])
word_counts_sum = word_counts.sum(axis=0)
words_freq = [(word, word_counts_sum[0, idx]) for word, idx in vec.vocabulary_.items()]
words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)
print(words_freq)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(x=sbrestaurants['date'].dt.year)
plt.title('Number of Tips Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Tips')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display summary statistics for the compliment_count column

print(sbrestaurants['compliment_count'].describe())

# Plot the distribution of compliment counts
plt.figure(figsize=(10, 6))
sns.countplot(x='compliment_count', data=tips)
plt.title('Distribution of Compliment Counts')
plt.xlabel('Compliment Count')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

top_businesses = sbrestaurants.groupby('business_id').size().sort_values(ascending=False).head(10)

top_businesses_with_names = top_businesses.to_frame().merge(sbrestaurants[['business_id', 'name']], on='business_id', how='left')

plt.figure(figsize=(10, 6))
plt.bar(top_businesses_with_names['name'], top_businesses_with_names[0], color='skyblue')
plt.title('Top 10 Businesses by Number of Tips')
plt.xlabel('Business Name')
plt.ylabel('Number of Tips')
plt.xticks(rotation=45)
plt.show()
```

```python
from textblob import TextBlob

sbrestaurants.loc[:, 'sentiment'] = sbrestaurants['text'].apply(lambda text: TextBlob(text).sentiment.polarity)

plt.figure(figsize=(10, 6))
sns.histplot(sbrestaurants['sentiment'], bins=30, kde=True)
plt.title('Sentiment Polarity Distribution')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

word_freq_dict = dict(words_freq)

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq_dict)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
